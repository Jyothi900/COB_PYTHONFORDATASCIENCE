```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv('netflix.csv')

#Count plot of show types
plt.figure(figsize=(10,6))
sns.countplot(x='type', data=dataset, palette='viridis')
plt.title('Count of Show Types on Netflix')
plt.xlabel('Show Type')
plt.ylabel('Count')
plt.show()
```

```python
top_countries = dataset['country'].value_counts().nlargest(10)
plt.figure(figsize=(12,6))
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Most Content on Netflix')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
top_directors = dataset['director'].value_counts().nlargest(10)
plt.figure(figsize=(12,6))
top_directors.plot(kind='bar', color='salmon')
plt.title('Top 10 Directors with Most Content on Netflix')
plt.xlabel('Director')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='rating', data=dataset, palette='rocket')
plt.title('Content Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
